```python
import re

text = "Hello, World! 123"

# Match words (A-Z, a-z)
print(re.findall(r"[A-Za-z]+", text))

# Match digits (0-9)
print(re.findall(r"\d+", text))

# Match special characters
print(re.findall(r"[^A-Za-z0-9 ]", text))
```

```python

text = "I have a cat and a dog."
print(re.findall(r"cat|dog", text))
```

```python
text = "<tag>content</tag><tag>more</tag>"

# Greedy (takes maximum)
print(re.findall(r"<tag>.*</tag>", text))

# Non-Greedy / Lazy (takes minimum)
print(re.findall(r"<tag>.*?</tag>", text))
```

```python
text = "gray grey"
print(re.findall(r"gr(a|e)y", text))
```

```python
text = "Line1\nLine2\tTabbed"
print(re.findall(r"\n", text))  # newline
print(re.findall(r"\t", text))  # tab
```

```python

print(re.match(r"Hello", "Hello World"))  # Starts with Hello
print(re.search(r"World$", "Hello World"))  # Ends with World
```

```python
text = "cat scatter cater"
print(re.findall(r"\bcat\b", text))   # Matches 'cat' alone
```

```python

text = "Price is $5.00"
print(re.findall(r"\$\d+\.\d+", text))
```

```python
text = "Name: Alice, Age: 21"
match = re.search(r"Name: (.*), Age: (\d+)", text)

print("Name:", match.group(1))
print("Age:", match.group(2))
```

```python
text = "Name: Alice, Age: 21"
match = re.search(r"Name: (.*), Age: (\d+)", text)

print("Name:", match.group(1))
print("Age:", match.group(2))
```

```python

text = "Python is fun and Python is powerful"
print(re.findall(r"Python(?= is)", text))  # Python followed by 'is'
```

```python
table = """
ID   Name   Marks
1    Ram    89
2    Shyam  92
3    Riya   85
"""

print(re.findall(r"[A-Za-z]+", table))  # Extract only names/words
```

```python
text = """start
middle
end"""

print(re.findall(r"start.*end", text, re.DOTALL))
```
