#### Question 
According to the U.S. Army Corps of Engineers, the top
20 U.S. ports, ranked by total tonnage (in million tons),
were as follows.


###### Discuss the shape of the distribution (skewness and Kurtosis) from the plot.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Given data
tonnage_values = [
    212.7, 211.7, 152.1, 83.9, 79.9, 78.9, 77.6, 65.9, 59.3, 57.8, 
    57.7, 54.9, 52.7, 49.2, 47.9, 44.7, 44.4, 44.1, 43.6, 39.4
]

# Compute statistics
mean_val = np.mean(tonnage_values)
median_val = np.median(tonnage_values)
std_dev = np.std(tonnage_values, ddof=1)
skewness = stats.skew(tonnage_values)
kurtosis = stats.kurtosis(tonnage_values)



# Determine skewness type
if skewness > 0:
    skew_type = "Positive Skewness (Right-skewed)"
elif skewness < 0:
    skew_type = "Negative Skewness (Left-skewed)"
else:
    skew_type = "Approximately Symmetric"

# Determine kurtosis type
if kurtosis > 3:
    kurtosis_type = "Leptokurtic (Heavy-tailed)"
elif kurtosis < 3:
    kurtosis_type = "Platykurtic (Light-tailed)"
else:
    kurtosis_type = "Mesokurtic (Normal distribution-like)"

# Plot histogram with density curve
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(tonnage_values, bins=10, kde=True, color="blue", stat="density")

# Add vertical lines for mean, media
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_val:.2f}')
plt.axvline(median_val, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_val:.2f}')

# Labels and legend
plt.xlabel("Total Tonnage (Million Tons)")
plt.ylabel("Density")
plt.title("Distribution of Total Tonnage of Top 20 U.S. Ports")
plt.legend()

# Show plot
plt.show()

# Print calculated values
print(f"Mean: {mean_val:.2f}")
print(f"Median: {median_val:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Skewness: {skewness:.2f} ({skew_type})")
print(f"Kurtosis: {kurtosis:.2f} ({kurtosis_type})")

# Explanation
print("\nSkewness Explanation:")
print("- Positive skewness means the right tail is longer, indicating a few high values are pulling the mean to the right.")
print("- Negative skewness means the left tail is longer, indicating a few low values are pulling the mean to the left.")
print("- If skewness is close to zero, the data is approximately symmetric.")

print("\nKurtosis Explanation:")
print("- Leptokurtic (>3) distributions have heavy tails and more extreme outliers.")
print("- Platykurtic (<3) distributions have light tails, meaning fewer extreme outliers.")
print("- Mesokurtic (3) distributions resemble a normal distribution.")
```
